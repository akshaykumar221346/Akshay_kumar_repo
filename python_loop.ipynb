1. ATM PIN Verification

```python
correct_pin = 1234
entered_pins = [1111, 2222, 1234]

attempts = 3

for pin in entered_pins[:3]:
    if pin == correct_pin:
        print("Access Granted")
        break
    else:
        attempts -= 1
        if attempts > 0:
            print(f"Wrong PIN, {attempts} tries left")
        else:
            print("Card Locked")
```

2. Grocery Checkout Total

```python
item_prices = [400, 350, 300, 200, 100]

total = 0
discount = False

for price in item_prices:
    if discount:
        total += price * 0.9
    else:
        total += price

        if total > 1000:
            discount = True

print("Final Total:", total)
```

3. Number Guessing Game

```python
secret_number = 25
guesses = [30, 20, 25, 18]

count = 0

for guess in guesses:
    count += 1

    if guess > secret_number:
        print("Too High")

    elif guess < secret_number:
        print("Too Low")

    else:
        print("Correct")
        print("Guesses Taken:", count)
        break
```

4. Movie Ticket Booking

```python
total_seats = 20
booking_requests = [5, 4, 6, 8]

available = total_seats

for request in booking_requests:

    if request <= available:
        available -= request
        print(f"{request} seats booked")

    else:
        print("Not enough seats")
        break

print("Seats Remaining:", available)
```

5. Study Streak Tracker

```python
scores = [75, 82, 90, 65, 59, 80]

streak = 0

for score in scores:

    if score >= 60:
        streak += 1

    else:
        print("Streak broken")
        break

print("Final Streak Length:", streak)
```
