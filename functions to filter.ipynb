Difference between print() and return

This is one of the most common interview questions.

print()
Displays the value on the screen.
Does not send the value back to the caller.
Cannot be reused later.

```python
def add(a, b):
    print(a + b)

x = add(10, 20)

print(x)
```

return
Sends the value back to whoever called the function.
Can be stored in a variable.
Can be used in another calculation.

```python
def add(a, b):
    return a + b

x = add(10, 20)

print(x)
print(x + 50)
```

```python
def square(num):
    return num ** 2

print(square(5))
```

```python
def is_prime(num):

    if num <= 1:
        return False

    for i in range(2, num):

        if num % i == 0:
            return False

    return True


number = 13

if is_prime(number):
    print("Prime Number")
else:
    print("Not Prime")
```

An iterator is an object that gives one value at a time.

Python remembers the current position.

Instead of returning everything together, it returns values one by one.

Example

```python
numbers = [10,20,30]

it = iter(numbers)

print(next(it))
print(next(it))
print(next(it))
```

```python
# ============================================================
# DRAWBACKS OF ITERATORS
# ============================================================

# 1. Moves only in the forward direction.

# 2. Once exhausted, it cannot be reused.
#    We must create a new iterator.

# Example

numbers = [1, 2]

it = iter(numbers)

print(next(it))
print(next(it))

# print(next(it))
# StopIteration Error
```

```python
# ============================================================
# GENERATORS
# ============================================================

# Definition:
# A generator is a special type of iterator.
# Instead of using return, it uses yield.
# It generates values one by one only when needed,
# making it memory efficient.

# yield
# -----
# • Pauses the function.
# • Remembers its current state.
# • Continues from where it stopped.

# return
# ------
# • Ends the function immediately.
# • Does not remember the previous state.

def numbers():

    yield 1
    yield 2
    yield 3

gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# ============================================================
# WHY GENERATORS?
# ============================================================

# Suppose we want numbers from 1 to 10 million.

# List
# ----
# Creates every number immediately.
# Consumes a lot of memory.

# Generator
# ---------
# Creates one number only when required.
# Saves memory.

def huge_numbers():

    for i in range(10000000):
        yield i
```

```python
# ============================================================
# FIBONACCI GENERATOR
# ============================================================

# Fibonacci Series
# 1 1 2 3 5 8 13 ...

def fibonacci(n):

    a = 1
    b = 1

    for i in range(n):

        yield a

        a, b = b, a + b


for num in fibonacci(10):
    print(num)
```

```python
# ============================================================
# MAP()
# ============================================================

# Definition:
# map() applies the same function to every element
# of an iterable.

# Syntax
# map(function, iterable)

numbers = [1,2,3,4,5]

def square(num):
    return num ** 2

result = list(map(square, numbers))

print(result)
```

```python
# ============================================================
# MAP EXAMPLE
# ============================================================

temps = [0,20,37,100]

def fahrenheit(c):
    return (9/5)*c + 32

result = list(map(fahrenheit, temps))

print(result)
```

```python
# ============================================================
# LAMBDA
# ============================================================

# Definition:
# A lambda function is an anonymous (nameless) function.
# It is mainly used when a function is needed only once.

# Syntax
# lambda arguments : expression

numbers = [1,2,3,4]

result = list(map(lambda x: x*x, numbers))

print(result)
```

```python
# ============================================================
# FILTER()
# ============================================================

# Definition:
# filter() returns only those values
# for which the condition is True.

# Syntax
# filter(function, iterable)

numbers = [1,2,3,4,5,6]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)
```

```python
# ============================================================
# REDUCE()
# ============================================================

# Definition:
# reduce() combines all values into one final value.
# It repeatedly applies a function to two elements.

# Unlike map() and filter(),
# reduce() returns only ONE value.

from functools import reduce

numbers = [1,2,3,4]

result = reduce(lambda x, y: x*y, numbers)

print(result)
```

```python
# ============================================================
# PRIME NUMBER USING FUNCTION
# ============================================================

# Definition:
# A prime number has exactly two factors:
# 1 and itself.

def is_prime(num):

    if num <= 1:
        return False

    for i in range(2, num):

        if num % i == 0:
            return False

    return True


number = 13

if is_prime(number):
    print("Prime Number")
else:
    print("Not Prime")
```

```python


# Function
# --------
# Reusable block of code.

# print()
# --------
# Displays output.

# return
# --------
# Sends value back and ends function.

# Iterator
# --------
# Returns one value at a time using iter() and next().

# Generator
# ---------
# Special iterator using yield.

# yield
# -----
# Pauses the function and remembers its state.

# map()
# -----
# Applies a function to every element.

# lambda
# -------
# Anonymous one-line function.

# filter()
# --------
# Keeps only values where the condition is True.

# reduce()
# --------
# Combines all values into one final result.
```
